```python
# Best case and worst case comparison for some algorithms.
import random
import time
import copy

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from insertion_sort import insertion_sort
from merge_sort import merge_sort
from quicksort import quicksort
```

```python
# Input for following usage.
input_list = {}
for i in range(1000, 21000, 1000):
    values = [random.uniform(0, 100000) for x in range(i)]
    input_list[i] = values
```

```python
# Insertion sort, inplace sort.
# Worst case: random input, O(n^2)
# Best case: sorted input, O(n)

insertion_input = copy.deepcopy(input_list)

worst_time_insertion = []
best_time_insertion = []

for length in insertion_input:
    # Worst case.
    start = time.time()
    insertion_sort(insertion_input[length])
    duration = time.time() - start
    worst_time_insertion.append([length, duration])
    # Best case.
    start = time.time()
    insertion_sort(insertion_input[length])
    duration = time.time() - start
    best_time_insertion.append([length, duration])
```

```python
# Plot insertion sort
line1, = plt.plot([x[0] for x in worst_time_insertion], [x[1] for x in worst_time_insertion], 'ro', label='Worst Case')
line2, = plt.plot([x[0] for x in best_time_insertion], [x[1] for x in best_time_insertion], 'co', label='Best Case')

plt.legend(handles=[line1, line2], loc=1)
plt.xlabel("Input Length")
plt.ylabel("Sorting Time (s)")
plt.title("Worst/Best Case Time Complexity for Insertion Sort")
plt.show()
```

```python
# Merge sort, inplace sort.
# Worst case: all input, O(nlgn)
# Best case: all input, O(nlgn)
# For natural merge sort, a variant of merge sort, the bese case is O(n)

merge_input = copy.deepcopy(input_list)

worst_time_merge = []
best_time_merge = []

for length in merge_input:
    # Worst case.
    start = time.time()
    merge_sort(merge_input[length], 0, length)
    duration = time.time() - start
    worst_time_merge.append([length, duration])
    # Best case.
    start = time.time()
    merge_sort(merge_input[length], 0, length)
    duration = time.time() - start
    best_time_merge.append([length, duration])
```

```python
# Plot merge sort
line1, = plt.plot([x[0] for x in worst_time_merge], [x[1] for x in worst_time_merge], 'ro', label='Worst Case')
line2, = plt.plot([x[0] for x in best_time_merge], [x[1] for x in best_time_merge], 'co', label='Best Case')

plt.legend(handles=[line1, line2], loc=1)
plt.xlabel("Input Length")
plt.ylabel("Sorting Time (s)")
plt.title("Worst/Best Case Time Complexity for Merge Sort")
plt.show()
```

```python
# Quicksort, inplace sort.
# Worst case: unbalanced input, e.g. small range of numbers, O(n^2)
# Best case: random input, O(nlgn)

import sys
sys.setrecursionlimit(10000)

quick_input_best = copy.deepcopy(input_list)
# Worst case: few unique numbers.
quick_input_worst = {}
for i in range(1000, 21000, 1000):
    values = [random.randint(0, 10) for x in range(i)]
    quick_input_worst[i] = values

worst_time_quick = []
best_time_quick = []

for length in quick_input_worst:
    # Worst case.
    start = time.time()
    quicksort(quick_input_worst[length], 0, length)
    duration = time.time() - start
    worst_time_quick.append([length, duration])
    # Best case.
    start = time.time()
    quicksort(quick_input_best[length], 0, length)
    duration = time.time() - start
    best_time_quick.append([length, duration])
```

```python
# Plot merge sort
line1, = plt.plot([x[0] for x in worst_time_quick], [x[1] for x in worst_time_quick], 'ro', label='Worst Case')
line2, = plt.plot([x[0] for x in best_time_quick], [x[1] for x in best_time_quick], 'co', label='Best Case')

plt.axis([0, 20000, 0, 0.1])

plt.legend(handles=[line1, line2], loc=1)
plt.xlabel("Input Length")
plt.ylabel("Sorting Time (s)")
plt.title("Worst/Best Case Time Complexity for Quick Sort")
plt.show()
```
